### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type

total_rides_city = pyber_data_df.groupby(["type"]).count()["ride_id"]

    #creates Series with index as elements of "type" (grouping by that column)
    #and as values in series, counts the number of "ride_id" and attributes it to each 
    # city type.
    
```

```python
# 2. Get the total drivers for each city type 

total_drivers_city = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_city

```

```python
#  3. Get the total amount of fares for each city type
total_fares_city = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_city
```

```python
#  4. Get the average fare per ride for each city type. 
#sum of all the fares / total number of rides for specific city 

avg_fare_rides_city = total_fares_city/total_rides_city


```

```python
# 5. Get the average fare per driver for each city type. 
#sum of all the fares / total drivers

avg_fare_drivers_city = total_fares_city/total_drivers_city

```

```python
#  6. Create a PyBer summary DataFrame. 

## merging 
#total_fares_city, total_drivers_city, total_rides_city
# pyber_summary_df = pd.merge(total_fares_city, total_drivers_city, total_rides_city)


pyber_summary_df = pd.DataFrame() # Create an empty dataframe!!!

#"pyber_summary_df" is a new df we make - the names in [] are the names of the column of new df
    #insert the series we made earlier to populate these columns
pyber_summary_df["Total Rides"] = total_rides_city
pyber_summary_df["Total Drivers"] = total_drivers_city
pyber_summary_df["Total Fares"] = total_fares_city
pyber_summary_df["Average Fare per Ride"] = avg_fare_rides_city
pyber_summary_df["Average Fare per Driver"] = avg_fare_drivers_city


```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns. 
    # this syntax is still tricky for me
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)




pyber_summary_df


```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame 
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.  --> so [type], [date]

fares_date_type = pyber_data_df.groupby(["date","type"]).sum()["fare"]
fares_date_type
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
fares_date_type = fares_date_type.reset_index()
fares_date_type
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

pyber_pivot = pd.pivot_table(fares_date_type, values="fare", index = "date", columns ="type") 
            #Creates pivot table from df "fares_date_type"
pyber_pivot



```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
pyber_loc_pivot = pyber_pivot.loc["2019-01-01":"2019-04-29"]
pyber_loc_pivot
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
pyber_loc_pivot.index = pd.to_datetime(pyber_loc_pivot.index)
pyber_loc_pivot

```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pyber_loc_pivot.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
#resample() --> weekly, monthly, etc. frequency


weekly_fare_sum = pyber_loc_pivot.resample("W").sum()
weekly_fare_sum

```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')  # what is this?

#fig, ax = plt.subplots() ---> why no this? 

weekly_fare_sum.plot(figsize=(25,8))
plt.ylabel("Fare($)")
plt.xlabel("Months")
plt.title("Total Fare by Type of City")
plt.legend(loc = "center")    ################### According to the module challenge, the legend should be in the middle of the graph?


plt.savefig("analysis/challenge_fig.png")





```
